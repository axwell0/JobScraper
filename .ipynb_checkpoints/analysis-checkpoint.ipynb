```python
import re
import asyncio
import string
import numpy as np
import pymongo
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import folium

from nltk.corpus import stopwords
from folium.plugins import HeatMap, MarkerCluster
from folium import Circle, Marker 
stop = set(stopwords.words('french'))
exclude = set(string.punctuation) 

from dotenv import load_dotenv
from utils.utils import get_job_industries
from utils.utils import get_coordinates

load_dotenv('database/.env')
bayt_collection_name = os.getenv('Bayt_collection')
tanit_collection_name = os.getenv('Tanit_collection')

MONGO_URI = os.getenv('MONGODB_URI')
DB_NAME = os.getenv('DATABASE_NAME')

client = pymongo.MongoClient(MONGO_URI)
db = client.get_database(DB_NAME)

tanit = db[tanit_collection_name]
bayt = db[bayt_collection_name]
```

```python
tanit_df = pd.DataFrame(iter(tanit.find()))
del tanit_df['_id']
tanit_df.rename(columns=
          {
              'Description de l\'emploi': 'Description',
              'Exigences de l\'emploi' : 'Exigences',
          }, inplace=True)
```

# Pre-Processing

- Extracting the number of available openings of a particular job from "Poste vacants" 

```python
job_openings = tanit_df['Postes vacants'].map(lambda poste: re.search(r'\d', poste).group()).astype('int32')
tanit_df['Postes vacants'] = job_openings
```

- Finding the Top employers on TanitJobs
<br>

```python
filtered_df = tanit_df[tanit_df['Employer'].str.lower() != 'anonyme']
top_employers = filtered_df.Employer.value_counts().head(10)
top_employers
```

```python
def clean(doc):
    if not(type(doc) in [np.nan,np.NAN,np.NaN,float,"",None]):
        stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
        punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
        only_alphabet = ''.join([c for c in punc_free if c.isalpha() or c == ' '])
        
        return only_alphabet
    return None


tanit_df['Description'] = tanit_df['Description'].fillna('').map(lambda x: clean(x)).apply(lambda x: x.lower())
tanit_df['Exigences'] = tanit_df['Exigences'].fillna('').map(lambda x: clean(x)).apply(lambda x: x.lower())
tanit_df['Title'].map(lambda x: x.encode('ascii',"ignore").decode())
```

```python
tanit_df['Zone'] = tanit_df['Zone'].map(lambda x: x.encode('ascii',"ignore").decode())
loop = asyncio.get_event_loop()
task = loop.create_task(get_coordinates(tanit_df))
coordinates = await task
tanit_df = tanit_df.merge(coordinates, on='Zone', how='left')
```

```python
tanit_df['Latitude'] = tanit_df['Latitude'].astype(float)
tanit_df['Longitude'] = tanit_df['Longitude'].astype(float) 
```

```python

loop = asyncio.get_event_loop()
task = loop.create_task(get_job_industries(tanit_df))
task_result = await task
tanit_df['Category'] = tanit_df['Title'].map(task_result)
```

```python
filtered_df = tanit_df[(tanit_df['Latitude'].notna()) & (tanit_df['Category'].notna())]
```

```python
m = folium.Map(location=[36.806389, 10.181667], zoom_start=6,tiles ='cartodbpositron')
mc = MarkerCluster()
for idx, row in filtered_df.iterrows():
    mc.add_child(Marker([row['Latitude'], row['Longitude']]))
m.add_child(mc)
```

```python

fig = px.scatter_mapbox(filtered_df, lat = 'Latitude', lon = 'Longitude',
                        color = 'Category',
                        zoom = 3, mapbox_style = 'open-street-map')
fig.show()
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(y=top_employers.index, x=top_employers.values)
plt.title('Top 10 Most Frequent Employers')
plt.xlabel('Employer')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
fig = px.bar(
    y=top_employers.index,
    x=top_employers.values,
    labels={'y': 'Employer', 'x': 'Frequency'},  # Adjusting axis labels
    title='Top 10 Most Frequent Employers',
    orientation='h'  # Horizontal bar plot
)

fig.update_layout(
    plot_bgcolor='black',  # Set background color to black
    paper_bgcolor='black',  # Set plot paper color to black
    font_color='white',  # Set font color to white
    yaxis=dict(tickfont=dict(size=12), automargin=True),  # Adjusting tick font size and setting automargin to True
    margin=dict(l=200),
    width=1000,  # Set the width of the figure
    height=400# Adding left margin to prevent labels from touching the axis line
)

fig.show()
```
